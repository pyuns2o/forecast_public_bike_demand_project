```python
weather = pd.read_csv('weather_modify.csv')
```

```python
weather
```

```python
weather['windspeed'].hist()
```

```python
weather['windforce'] = 0
weather.loc[weather['windspeed'] >= 29, 'windforce'] = 1
weather.head()
```

```python
def humidex_cal(T, Td):
    upper = 5417.7530*(1/273.16 - 1/(273.15+Td))
    inner = 6.11 * (2.71828 **upper)
    return  T + 0.5555*(inner - 10)
```

```python
humidex_cal(weather['temp'], weather['dew'])
```

```python
weather['humidex'] = humidex_cal(weather['temp'], weather['dew'])
weather.head()
```

```python
weather['discomfort'] = 0
weather.loc[weather['humidex'] >= 30, 'discomfort'] = 1
weather.loc[weather['humidex'] >= 40, 'discomfort'] = 2
weather.loc[weather['humidex'] >= 46, 'discomfort'] = 3
weather.head()
# 숫자가 올라갈수록 불쾌
```

```python
def HI_cal(T,RH):
    HI = -42.379 + 2.04901523*T + 10.14333127*RH - .22475541*T*RH - .00683783*T*T - .05481717*RH*RH + .00122874*T*T*RH + .00085282*T*RH*RH - .00000199*T*T*RH*RH
    return HI
```

```python
weather['화씨'] = weather['temp'] *(9/5) + 32
```

```python
weather['HI_화씨'] = HI_cal(weather['화씨'], weather['humid'])
```

```python
weather['HI_섭씨'] = (5/9)*(weather['HI_화씨'] - 32)
```
